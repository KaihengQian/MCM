```python
#处理特征工程
import csv

#指定比赛场次
Match_id = "2023-wimbledon-1301"
#指定窗口大小
window_size = 5 

match = 1

match_data_list = []


with open('Wimbledon_featured_matches_washed.csv', 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader, None)
    
    for row in csv_reader:
        if row[match]==Match_id:
            match_data_list.append(row)
          
#处理数据，原来全部都是字符型变量
for row in match_data_list:
    for i in range(4, 12):  # 第4-11列
        row[i] = float(row[i]) 
    for i in range(14, len(row)):  # 第16-最后一列
        if row[i] == '-1' or row[i] == '':
            row[i] = 0
        row[i] = float(row[i])
        
#print(match_data_list[1])
```

```python
#计算p1各项特征
point_no = 7
p1_sets = 8
p2_sets = 9
p1_games = 10
p2_games = 11
server = 14
server_no = 15
point_victor = 16
p1_ace = 21
p2_ace = 22
p1_winner = 23
p2_winner = 24
winner_shot_type =25
p1_double_fault = 26
p2_double_fault = 27
p1_unf_err = 28
p2_unf_err = 29
p1_net_pt = 30
p2_net_pt = 31
p1_net_pt_won = 32
p2_net_pt_won = 33
p1_break_pt = 34
p2_break_pt = 35
p1_break_pt_won = 36
p2_break_pt_won = 37
p1_distance_run = 40
p2_distance_run = 41
rally_count = 42
speed_mph = 43
serve_width = 44
serve_depth = 45
return_depth =46


point_number = 0
p1_sets_to_win = 1
p1_games_to_win = 2
p2_sets_to_win = 1
p2_games_to_win = 2
first_serve_success_rate = 3
point_victor_rate = 4
p1_ace_sum = 5
p1_winner_sum = 6
p1_double_fault_sum = 7
p1_unf_err_sum = 8
p1_net_pt_won_rate = 9
p1_break_pt_won_rate = 10
p1_rally_distance_avg = 11
p2_ace_sum = 5
p2_winner_sum = 6
p2_double_fault_sum = 7
p2_unf_err_sum = 8
p2_net_pt_won_rate = 9
p2_break_pt_won_rate = 10
p2_rally_distance_avg = 11
speed_mph_avg =12
serve_depth_avg = 13
return_depth_avg = 14
serve_width_avg = 15
p1_winner_type_rate = 16
p2_winner_type_rate = 16

def calculate_p1(current_point_number):
    p1_data_given_time = [None] * 17
    
    player_no = 1
    pn =  int(current_point_number)-1
    first_serve_success = 0
    serve_number = 0
    victor_number = 0
    ace_number = 0
    winner_number = 0
    double_fault_number = 0
    unf_err_number = 0
    net_pt_number = 0
    net_pt_won_number = 0
    break_pt_number = 0
    break_pt_won_number = 0
    distance_run_number = 0
    rally_count_number = 0
    speed_mph_number = 0
    serve_depth_number = 0
    return_depth_number = 0
    serve_width_number = 0
    winner_shot_type_number = 0
    
    
    p1_data_given_time[point_number] = current_point_number
    p1_data_given_time[p1_sets_to_win] = 3 + min(match_data_list[pn][p1_sets],match_data_list[pn][p2_sets]) - match_data_list[pn][p1_sets]
    if match_data_list[pn][p2_games] < 5:
        p1_data_given_time[p1_games_to_win] = 6 - match_data_list[pn][p1_games]
    else:
        p1_data_given_time[p1_games_to_win] = 7 - match_data_list[pn][p1_games]
        
    
    for i in range(window_size):
        if match_data_list[pn-i][server] == player_no:
            serve_number += 1
            if match_data_list[pn-i][server_no] == 1:
                first_serve_success += 1
            serve_depth_number += match_data_list[pn-i][serve_depth]
            serve_width_number += match_data_list[pn-i][serve_width]
        else:
            return_depth_number += match_data_list[pn-i][return_depth]
        
        if match_data_list[pn-i][point_victor] == player_no:
            victor_number += 1
            
        if match_data_list[pn-i][p1_ace] == 1:
            ace_number += 1
            
        if match_data_list[pn-i][p1_winner] == 1:
            winner_number += 1
            winner_shot_type_number += match_data_list[pn-i][winner_shot_type]
            
        if match_data_list[pn-i][p1_double_fault] == 1:
            double_fault_number += 1
    
        if match_data_list[pn-i][p1_unf_err] == 1:
            unf_err_number += 1
            
        if match_data_list[pn-i][p1_net_pt] == 1:
            net_pt_number += 1
            
        if match_data_list[pn-i][p1_net_pt_won] == 1:
            net_pt_won_number += 1
            
        if match_data_list[pn-i][p1_break_pt] == 1:
            break_pt_number += 1
        
        if match_data_list[pn-i][p1_break_pt_won] == 1:
            break_pt_won_number += 1
            
            
        distance_run_number += match_data_list[pn-i][p1_distance_run]
        rally_count_number += match_data_list[pn-i][rally_count]
        speed_mph_number += match_data_list[pn-i][speed_mph]
        
            
    if serve_number == 0:
        p1_data_given_time[first_serve_success_rate] = 0
        p1_data_given_time[serve_depth_avg] = 0
        p1_data_given_time[serve_width_avg] = 0
        p1_data_given_time[return_depth_avg] = return_depth_number / window_size
    else:
        p1_data_given_time[first_serve_success_rate] = first_serve_success / serve_number
        p1_data_given_time[serve_depth_avg] = serve_depth_number / serve_number
        p1_data_given_time[serve_width_avg] = serve_width_number / serve_number
        if window_size - serve_number == 0:
            p1_data_given_time[return_depth_avg] = 0
        else:
            p1_data_given_time[return_depth_avg] = return_depth_number / (window_size - serve_number)
        
        
    p1_data_given_time[point_victor_rate] = victor_number / window_size
    p1_data_given_time[p1_ace_sum] = ace_number
    p1_data_given_time[p1_winner_sum] = winner_number
    p1_data_given_time[p1_double_fault_sum] = double_fault_number
    p1_data_given_time[p1_unf_err_sum] = unf_err_number
    
    if net_pt_number == 0:
        p1_data_given_time[p1_net_pt_won_rate] = 0
    else:
        p1_data_given_time[p1_net_pt_won_rate] = net_pt_won_number / net_pt_number
    
    if break_pt_number == 0:
        p1_data_given_time[p1_break_pt_won_rate] = 0
    else:
        p1_data_given_time[p1_break_pt_won_rate] = break_pt_won_number / break_pt_number 
    
    p1_data_given_time[p1_rally_distance_avg] = distance_run_number / (rally_count_number+1)
    p1_data_given_time[speed_mph_avg] = speed_mph_number / window_size
    
    if winner_number == 0:
        p1_data_given_time[p1_winner_type_rate] = 0
    else:
        p1_data_given_time[p1_winner_type_rate] = winner_shot_type_number / winner_number
    
    
    #print(p1_data_given_time)
    return p1_data_given_time

def calculate_p2(current_point_number):
    p2_data_given_time = [None] * 17
    
    player_no = 2
    pn =  int(current_point_number)-1
    first_serve_success = 0
    serve_number = 0
    victor_number = 0
    ace_number = 0
    winner_number = 0
    double_fault_number = 0
    unf_err_number = 0
    net_pt_number = 0
    net_pt_won_number = 0
    break_pt_number = 0
    break_pt_won_number = 0
    distance_run_number = 0
    rally_count_number = 0
    speed_mph_number = 0
    serve_depth_number = 0
    return_depth_number = 0
    serve_width_number = 0
    winner_shot_type_number = 0
    
    
    p2_data_given_time[point_number] = current_point_number
    p2_data_given_time[p1_sets_to_win] = 3 + min(match_data_list[pn][p2_sets],match_data_list[pn][p1_sets]) - match_data_list[pn][p2_sets]
    if match_data_list[pn][p1_games] < 5:
        p2_data_given_time[p2_games_to_win] = 6 - match_data_list[pn][p2_games]
    else:
        p2_data_given_time[p2_games_to_win] = 7 - match_data_list[pn][p2_games]
        
    
    for i in range(window_size):
        if match_data_list[pn-i][server] == player_no:
            serve_number += 1
            if match_data_list[pn-i][server_no] == 1:
                first_serve_success += 1
            serve_depth_number += match_data_list[pn-i][serve_depth]
            serve_width_number += match_data_list[pn-i][serve_width]
        else:
            return_depth_number += match_data_list[pn-i][return_depth]
        
        if match_data_list[pn-i][point_victor] == player_no:
            victor_number += 1
            
        if match_data_list[pn-i][p2_ace] == 1:
            ace_number += 1
            
        if match_data_list[pn-i][p2_winner] == 1:
            winner_number += 1
            winner_shot_type_number += match_data_list[pn-i][winner_shot_type]
            
        if match_data_list[pn-i][p2_double_fault] == 1:
            double_fault_number += 1
    
        if match_data_list[pn-i][p2_unf_err] == 1:
            unf_err_number += 1
            
        if match_data_list[pn-i][p2_net_pt] == 1:
            net_pt_number += 1
            
        if match_data_list[pn-i][p2_net_pt_won] == 1:
            net_pt_won_number += 1
            
        if match_data_list[pn-i][p2_break_pt] == 1:
            break_pt_number += 1
        
        if match_data_list[pn-i][p2_break_pt_won] == 1:
            break_pt_won_number += 1
            
            
        distance_run_number += match_data_list[pn-i][p2_distance_run]
        rally_count_number += match_data_list[pn-i][rally_count]
        speed_mph_number += match_data_list[pn-i][speed_mph]
        
            
    if serve_number == 0:
        p2_data_given_time[first_serve_success_rate] = 0
        p2_data_given_time[serve_depth_avg] = 0
        p2_data_given_time[serve_width_avg] = 0
        p2_data_given_time[return_depth_avg] = return_depth_number / window_size
    else:
        p2_data_given_time[first_serve_success_rate] = first_serve_success / serve_number
        p2_data_given_time[serve_depth_avg] = serve_depth_number / serve_number
        p2_data_given_time[serve_width_avg] = serve_width_number / serve_number
        if window_size - serve_number == 0:
            p2_data_given_time[return_depth_avg] = 0
        else:
            p2_data_given_time[return_depth_avg] = return_depth_number / (window_size - serve_number)
        
        
    p2_data_given_time[point_victor_rate] = victor_number / window_size
    p2_data_given_time[p1_ace_sum] = ace_number
    p2_data_given_time[p1_winner_sum] = winner_number
    p2_data_given_time[p1_double_fault_sum] = double_fault_number
    p2_data_given_time[p1_unf_err_sum] = unf_err_number
    
    if net_pt_number == 0:
        p2_data_given_time[p1_net_pt_won_rate] = 0
    else:
        p2_data_given_time[p1_net_pt_won_rate] = net_pt_won_number / net_pt_number
    
    if break_pt_number == 0:
        p2_data_given_time[p1_break_pt_won_rate] = 0
    else:
        p2_data_given_time[p1_break_pt_won_rate] = break_pt_won_number / break_pt_number 
    
    p2_data_given_time[p1_rally_distance_avg] = distance_run_number / (rally_count_number+1)
    p2_data_given_time[speed_mph_avg] = speed_mph_number / window_size
    
    if winner_number == 0:
        p2_data_given_time[p1_winner_type_rate] = 0
    else:
        p2_data_given_time[p1_winner_type_rate] = winner_shot_type_number / winner_number
    
    
    #print(p2_data_given_time)
    return p2_data_given_time
```
